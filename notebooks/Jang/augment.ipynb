```python
import random

def convert_5_to_4_choices(row):
    choices = eval(row['choices'])  # 리스트로 변환
    answer = row['answer']  # 1-based index
    
    if len(choices) != 5:
        return row  # 이미 4지선다면 그대로
    
    # 정답이 아닌 인덱스들
    wrong_indices = [i for i in range(5) if i != (answer - 1)]
    
    # 랜덤하게 1개 제거할 오답 선택
    remove_idx = random.choice(wrong_indices)
    
    # 새로운 choices 생성
    new_choices = [choices[i] for i in range(5) if i != remove_idx]
    
    # answer 조정 (제거된 선택지가 정답보다 앞에 있으면 -1)
    new_answer = answer if remove_idx >= (answer - 1) else answer - 1
    
    return new_choices, new_answer
```

```python
import random
import pandas as pd
import ast
from collections import Counter

sat_df_original = pd.read_csv(project_root / "data/train_augment_SAT.csv")

def convert_5_to_4_with_seed(seed_value):
    """특정 seed로 변환하고 answer 분포 반환"""
    random.seed(seed_value)
    
    sat_df = sat_df_original.copy()
    
    def convert_5_to_4_choices(problems_str):
        problems = ast.literal_eval(problems_str)
        choices = problems['choices']
        answer = problems['answer']
        
        if len(choices) != 5:
            return problems_str
        
        wrong_indices = [i for i in range(5) if i != (answer - 1)]
        remove_idx = random.choice(wrong_indices)
        new_choices = [choices[i] for i in range(5) if i != remove_idx]
        new_answer = answer if remove_idx >= (answer - 1) else answer - 1
        
        new_problems = {
            'question': problems['question'],
            'choices': new_choices,
            'answer': new_answer
        }
        return str(new_problems)
    
    sat_df['problems'] = sat_df['problems'].apply(convert_5_to_4_choices)
    
    # answer 분포 계산
    answers = sat_df['problems'].apply(lambda x: ast.literal_eval(x)['answer']).tolist()
    distribution = Counter(answers)
    
    return sat_df, distribution

# 여러 seed 시도
print("=== Seed별 Answer 분포 탐색 ===\n")

best_seed = None
best_std = float('inf')
best_dist = None

for seed in range(100):  # 0~99까지 시도
    _, dist = convert_5_to_4_with_seed(seed)
    
    # 표준편차 계산 (균등도 측정)
    counts = [dist.get(i, 0) for i in range(1, 5)]
    mean = sum(counts) / len(counts)
    std = (sum((x - mean)**2 for x in counts) / len(counts)) ** 0.5
    
    if std < best_std:
        best_seed = seed
        best_std = std
        best_dist = dist
    
    if seed < 10:  # 처음 10개만 출력
        print(f"Seed {seed:2d}: {dict(sorted(dist.items()))} (std: {std:.2f})")

print(f"\n{'='*50}")
print(f"✅ 최적 Seed: {best_seed}")
print(f"   분포: {dict(sorted(best_dist.items()))}")
print(f"   표준편차: {best_std:.2f}")
print(f"   이상적 분포 (26.75개씩): {dict(zip(range(1, 5), [26.75]*4))}")

# 최적 seed로 최종 변환
sat_df_final, _ = convert_5_to_4_with_seed(best_seed)
sat_df_final.to_csv(project_root / "data/train_augment_SAT_4.csv", index=False)
print(f"\n✅ train_augment_SAT_4.csv 저장 완료! (seed={best_seed})")
```

```python
# 원본과 변환된 데이터 로드
sat_df_original = pd.read_csv(project_root / "data/train_augment_SAT.csv")
sat_df_4 = pd.read_csv(project_root / "data/train_augment_SAT_4.csv")

# problems 파싱
sat_df_original['problems_parsed'] = sat_df_original['problems'].apply(ast.literal_eval)
sat_df_4['problems_parsed'] = sat_df_4['problems'].apply(ast.literal_eval)

# 검증: 각 행마다 정답 텍스트가 동일한지 확인
print("=== 전체 데이터 검증 ===\n")
random.seed(42)
errors = []
for idx in range(len(sat_df_original)):
    orig = sat_df_original.iloc[idx]['problems_parsed']
    conv = sat_df_4.iloc[idx]['problems_parsed']
    
    # 원본 정답 텍스트
    orig_answer_text = orig['choices'][orig['answer'] - 1]
    
    # 변환된 정답 텍스트
    conv_answer_text = conv['choices'][conv['answer'] - 1]
    
    # 검증
    if orig_answer_text != conv_answer_text:
        errors.append({
            'idx': idx,
            'id': sat_df_original.iloc[idx]['id'],
            'orig_answer': orig['answer'],
            'orig_text': orig_answer_text,
            'conv_answer': conv['answer'],
            'conv_text': conv_answer_text,
            'orig_choices': orig['choices'],
            'conv_choices': conv['choices']
        })

if errors:
    print(f"오류 발견! {len(errors)}개 불일치\n")
    for err in errors[:5]:  # 처음 5개만 출력
        print(f"Row {err['idx']} (ID: {err['id']}):")
        print(f"  원본: {err['orig_answer']}번 - {err['orig_text']}")
        print(f"  변환: {err['conv_answer']}번 - {err['conv_text']}")
        print(f"  원본 choices: {err['orig_choices']}")
        print(f"  변환 choices: {err['conv_choices']}\n")
else:
    print(f"검증 완료! 총 {len(sat_df_original)}개 모두 정답 일치")

# 추가 통계
print(f"\n=== 통계 ===")
print(f"원본 choices 개수: {sat_df_original['problems_parsed'].apply(lambda x: len(x['choices'])).value_counts().to_dict()}")
print(f"변환 choices 개수: {sat_df_4['problems_parsed'].apply(lambda x: len(x['choices'])).value_counts().to_dict()}")
print(f"\n원본 answer 분포: {sat_df_original['problems_parsed'].apply(lambda x: x['answer']).value_counts().sort_index().to_dict()}")
print(f"변환 answer 분포: {sat_df_4['problems_parsed'].apply(lambda x: x['answer']).value_counts().sort_index().to_dict()}")
```
