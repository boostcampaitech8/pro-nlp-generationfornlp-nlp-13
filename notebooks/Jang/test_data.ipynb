```python
def make_predictoin_df(file_path, test_df):
    df = pd.read_csv(project_root / file_path)
    df_r = df.rename(columns={'answer':'predict'})
    predict_df = pd.merge(test_df, df_r, on='id', how='left')

    return predict_df
```

```python
predict_df_sub_dpo = make_predictoin_df('submission_dpo.csv', test_df)
predict_df_sub_2 =  make_predictoin_df('submission2.csv', test_df)
predict_df_sub_3 =  make_predictoin_df('submission3.csv', test_df)
predict_df_sub_4 =  make_predictoin_df('submission4.csv', test_df)
predict_df_sub_5 =  make_predictoin_df('submission5.csv', test_df)
```

```python
predict_list = [predict_df_sub_dpo, predict_df_sub_2, predict_df_sub_3, predict_df_sub_4, predict_df_sub_5]
names = ["DPO", "Model_2", "Model_3", "Model_4", "Model_5"]

for i,df in enumerate(predict_list):
    print("="*70)
    print(f"predict Counts: {names[i]}")
    counts = df['predict'].value_counts(dropna=False)
    probs = df['predict'].value_counts(dropna=False, normalize=True) * 100
    
    result = pd.concat([counts, probs.round(1)], axis=1)
    result.columns = ['Count', 'Percentage(%)']
    print(result)

```

```python
for i, df in enumerate(predict_list):
    print("="*70)
    print(f"MODEL: {names[i]}")
    counts = df['predict'].value_counts(dropna=False)
    probs = df['predict'].value_counts(dropna=False, normalize=True) * 100
    
    result = pd.concat([counts, probs.round(1)], axis=1)
    result.columns = ['Count', 'Percentage(%)']
    print(result)

    if 'choices_count' not in df.columns:
        df['choices_count'] = df['choices'].apply(len)
    
    for count in sorted(df['choices_count'].unique()):
        subset = df[df['choices_count'] == count]
        
        print(f"\n--- {count}-지선다 문제 (Total: {len(subset)}개) ---")
        
        counts = subset['predict'].value_counts(dropna=False).sort_index()
        probs = (subset['predict'].value_counts(dropna=False, normalize=True) * 100).round(1)
        
        result = pd.concat([counts, probs], axis=1)
        result.columns = ['Count', 'Percentage(%)']
        print(result)
```
